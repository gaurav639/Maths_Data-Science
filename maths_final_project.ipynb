```python
import io
import pandas as pd
```

```python
df = pd.read_excel(io.BytesIO(uploaded['TLP_LeaderProject_Data.xlsx'])).fillna(0)
df
```

```python
import numpy as np
```

```python
arr = df.to_numpy()[1:][:,2:]
```

```python
arr
```

```python
import random

# We created a empty list where we will keep the counter of which member is reached by any other member how many times
values_list = [0] * 42

#First we have taken a random index 
random_index = random.randint(0,41)

for i in range(100000):

  # Now we are storing the column based on the above random index
  temp = np.array(arr[:,random_index])

  # We again take a random index to select one element again from the above selected column
  random_index = random.randint(0,41)
  
  # This condition checks if somebody marked that person at the random index of temp as a leader then its value inside value_list increased to 1
  if temp[random_index] == 1:
    values_list[random_index] = values_list[random_index] + 1

  # If random index comes 0 then we again choose another random index and run the process
  else:
    random_index = random.randint(0,41)

print(values_list)  
```

```python
names = df.to_numpy()[1:,1]

distribution = {}

for a, b in zip(names, values_list):
  distribution[a] = b

print(distribution)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 3))
plt.bar(distribution.keys(),distribution.values())

plt.xticks(rotation = 90)

plt.grid()
plt.show()
```
